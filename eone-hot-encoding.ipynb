```python
import pandas as pd
import numpy as np

data = {'Letters': ['A', 'B', 'A', 'C', 'B'],
        'Colors': ['red','blue','pink','red','blue']}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
def one_hot_encoding(data, column):
    unique_values = data[column].unique()
    dummy_columns = {}

    for value in unique_values:
        dummy_column = f"{column}_{value}"
        data[dummy_column] = np.where(data[column] == value, 1, 0)
        dummy_columns[value] = dummy_column

    return data.drop(column, axis=1), dummy_columns



encoded_df, dummy_columns = one_hot_encoding(df, 'Letters')
encoded_df1, dummy_columns1 = one_hot_encoding(encoded_df, 'Colors')



print("\nDataFrame with Custom get_dummies:")
print(encoded_df1)

print("\nDummy Columns:")
print(dummy_columns,dummy_columns1)
```
